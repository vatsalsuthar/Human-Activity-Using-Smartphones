```python
def feature_normalize(df):
    mu = np.mean(df,axis = 0)
    sigma = np.std(df,axis = 0)
    return (df - mu)/sigma
```

```python
def plot_axis(ax, x, y, title):
    ax.plot(x, y)
    ax.set_title(title)
    ax.xaxis.set_visible(False)
    ax.set_ylim([min(y) - np.std(y), max(y) + np.std(y)])
    ax.set_xlim([min(x), max(x)])
    ax.grid(True)
```
